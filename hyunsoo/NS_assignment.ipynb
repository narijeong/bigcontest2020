```python
from scipy.optimize import linear_sum_assignment
import numpy as np
import pandas as pd


# 데이터 불러오기 
sales = pd.read_excel('NS_time.xlsx')
sales.head()
```

# 데이터 전처리

```python
# 데이터 전처리 전 row 개수 
print(len(sales))

# 데이터 전처리 전 column 개수
print(len(sales.columns))
```

```python
# 중복행 제거하기

sales = sales.drop_duplicates(['Unnamed: 0'], keep='first')

# 최종 row 개수 
len(sales)
```

```python
# input data 형식으로 바꿔주기 - 방송 시간 column 제외 

sales_df = sales.iloc[0:,1:]
```

```python
# column 추가하기 - row 개수 이상으로 column 개수 맞추기 

sales_add = pd.concat([sales_df, sales_df,sales_df,sales_df,sales_df,sales_df], axis=1)
sales_add.head()
```

```python
# input data 형식으로 바꿔주기

sales_matrix = np.array(sales_add)
sales_matrix
```

# linear_sum_assignment
- Parameters
- maximize : bool (default: False)
- Calculates a maximum weight matching if true.


```python
# 할당 알고리즘에 넣기

row_ind, col_ind = linear_sum_assignment(sales_matrix, True)

# col_ind : 시간대별 가장 높은 매출 가져오는 상품의 인덱스

col_ind
```

```python
# row 개수와 인덱스 개수 일치 여부 확인 

len(col_ind)
```

# 상품 편성 및 예측 매출 df 만들기

```python
# 상품명 list 
lists = sales_add.columns
lists
```

```python
# 최적화 편성 상품 데이터 

series1 = [lists[x] for x in col_ind] 
series1
```

```python
# 최적화 편성 취급액 데이터
series2 = [sales_add.iloc[x, col_ind[x]] for x in range(len(sales_add))]
series2
```

```python
# 최종 df 만들기 

# 시간 가져오기 
final_df = pd.DataFrame(sales.iloc[0:,0])

# 상품명 column 추가 
final_df['name'] = series1

# 취급액 column 추가 
final_df['sales'] = series2

final_df
```
